```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset("tips")


# Create separate dataframes for lunch and dinner

lunch_data = tips[tips["time"] == "Lunch"]
dinner_data = tips[tips["time"] == "Dinner"]


```

lunch_data is created by filtering the tips DataFrame to include only rows where the "time" column is "Lunch". This is done using boolean indexing: tips["time"] == "Lunch" returns a boolean Series that is True for rows where "time" is "Lunch" and False otherwise. Using this Series to index the tips DataFrame selects only the rows corresponding to lunch.


Similarly, dinner_data is created by filtering the tips DataFrame to include only rows where the "time" column is "Dinner".

```python
lunch_data.head()
```

```python
dinner_data.head()
```

```python
print(lunch_data.shape)
print(dinner_data.shape)
```

```python


# Set up the figure with two subplots side by side
fig, axes = plt.subplots(1, 2, figsize=(12, 5))     # creates a figure (fig) with two subplots (axes), arranged in a single row (1) with two columns (2). The figsize parameter sets the size of the figure to 12 inches by 5 inches.



# Plot histogram for lunch data
sns.histplot(lunch_data["total_bill"], kde=True, color='blue', bins=40, ax=axes[0])                                  
mean_lunch = lunch_data["total_bill"].mean()                                                                         
axes[0].axvline(mean_lunch, color='red', linestyle='dashed', linewidth=2, label=f"Mean: {mean_lunch:.2f}")
axes[0].set_xlabel("Total Bill")
axes[0].set_ylabel("Frequency")
axes[0].set_title("Distribution of Total Bill - Lunch")
axes[0].legend() 



# Plot histogram for dinner data
sns.histplot(dinner_data["total_bill"], kde=True, color='orange', bins=40, ax=axes[1])
mean_dinner = dinner_data["total_bill"].mean()
axes[1].axvline(mean_dinner, color='red', linestyle='dashed', linewidth=2, label=f"Mean: {mean_dinner:.2f}")
axes[1].set_xlabel("Total Bill")
axes[1].set_ylabel("Frequency")
axes[1].set_title("Distribution of Total Bill - Dinner")
axes[1].legend()



#Show the plot
plt.tight_layout() #The tight_layout function adjusts the subplots to fit into the figure area neatly, preventing overlap.
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Apply the default theme
sns.set_theme()      #This line sets the default aesthetic parameters for Seaborn plots, making them look better by default.

penguins = sns.load_dataset("penguins")

# Get the feature names (column names) from the DataFrame
feature_names = penguins.columns[:-1]  # Exclude the last column ("species")

# Get the unique class names (species)
class_names = penguins["species"].unique()

# Print the feature names and class names
print("Feature Names:", feature_names)
print("Class Names:", class_names)
```

```python
# Plot the class distribution
plt.figure(figsize=(8, 6))
sns.countplot(data=penguins, x="species")
plt.xlabel('Species')
plt.ylabel('Count')
plt.title('Class Distribution of Penguins Species')
plt.show()
```

sns.countplot is a Seaborn function that creates a bar plot showing the count of observations in each categorical bin using bars.

data=penguins specifies the DataFrame to be used.
x="species" sets the x-axis to represent the "species" column of the DataFrame, where each bar represents a different species of penguin.

```python
# Display information about the dataset
print(penguins.info())
```

numeric_columns = penguins.select_dtypes(include=[np.number]): This line selects only the numeric columns from the penguins DataFrame. The select_dtypes method is used to include columns with numeric data types.


cov_matrix = numeric_columns.cov(): This line calculates the covariance matrix for the numeric columns. The covariance matrix shows how much the variables change together.


variance_values = numeric_columns.var(): This line calculates the variance for each numeric column. The variance measures the spread of the data in each variable.

sns.pairplot: This Seaborn function creates a grid of scatter plots for each pair of numeric columns in the DataFrame, with different colors for each species.

data=penguins: Specifies the DataFrame to be used.

hue="species": Sets the color of the points in the scatter plots based on the "species" column, allowing for differentiation between species.

diag_kind="kde": Specifies that the diagonal plots should be kernel density estimates (KDE) plots instead of histograms.

```python
# Calculate covariance matrix and variance for each variable

numeric_columns = penguins.select_dtypes(include=[np.number])
cov_matrix = numeric_columns.cov()
variance_values = numeric_columns.var()

# Create the pairplot with covariance plots
g = sns.pairplot(data=penguins, hue="species", diag_kind="kde")


plt.show()
```
